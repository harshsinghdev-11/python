```python
from transformers import GemmaTokenizer
```

```python
os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
```

```python
model_id = "google/gemma-2b-it"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id,token=os.environ["HF_TOKEN"])
```

```python
model = AutoModelForCausalLM.from_pretrained(model_id,
                                             quantization_config=bnb_config,
                                             device_map={"":0},
                                             token=os.environ["HF_TOKEN"])
```

```python
text = "quote:imagination is more"
device = "cuda:0"
inputs = tokenizer(text, return_tensors="pt").to(device)
outputs = model.generate(**inputs, max_new_tokens=50)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
lora_config = LoraConfig(
    r=8,
    target_modules = ["q_proj","o_proj","k_proj","v_proj","gate_proj","up_proj",
                      "down_proj"],
    task_type="CAUSAL_LM"
)
```

```python
import kagglehub

# Download latest version
data_path = kagglehub.dataset_download("thedevastator/amod-mental-health-counseling-conversations-data")

print("Path to dataset files:", data_path)
```

```python
csv_files = [f for f in os.listdir(data_path) if f.endswith('.csv')]
print(csv_files)
```

```python
df = pd.read_csv(os.path.join(data_path,csv_files[0]))
print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(df.head())
```

```python
from datasets import Dataset
```

```python
train_dataset = Dataset.from_pandas(df)
```

```python
def formatting_func(example):
    text = f"<bos><start_of_turn>user\n{example['Context']}<end_of_turn>\n<start_of_turn>model\n{example['Response']}<end_of_turn><eos>"
    return text
```

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=100,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit"
    ),
    peft_config=lora_config,
    formatting_func = formatting_func
)
```

```python
trainer.train()
```

```python
context = "I'm going through some things with my feelings and myself. I barely sleep and I do nothing but think..."
device = "cuda:0"
inputs = tokenizer(context, return_tensors="pt").to(device)
outputs = model.generate(**inputs, max_new_tokens=100)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
